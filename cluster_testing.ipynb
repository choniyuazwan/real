```python
import pandas as pd
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.externals import joblib
```

```python
df_original = pd.read_excel('dummy original.xlsx', sheet_name='CLUSTERING')
```

```python
df = df_original
```

```python
df.head()
```

```python
df.shape
```

```python
# dapatkan data cardno, simpan ke dalam variabel cardno
cardno = df['CARDNO']
```

```python
# hapus kolom obs dan cardno dari df
df = df.drop(df.columns[[0,1]], axis=1)
```

```python
df.head()
```

```python
# isi data kosong atau null dengan nilai 0
df.fillna(0, inplace=True)
```

```python
# isi data . dengan nilai 0
df = df.replace({'.':0})
```

```python
df.head()
```

```python
# convert tipe data kolom2 pada df menjadi numerik
df = df.apply(pd.to_numeric)
```

```python
# simpan data nama kolom2 pada df kedalam variabel column_name
column_name = df.columns
```

```python
# load model scaler yang didapatkan dari cluster_training
scaler = joblib.load('model_scaler.pkl')
```

```python
# normalisasi nilai df menggunakan scaler
scaled = scaler.transform(df)
```

```python
scaled.shape
```

```python
# load model kmeans yang didapatkan dari cluster_training
kmeans = joblib.load('model_kmeans.pkl')
```

```python
# predict hasil cluster dari scaled
clusters = kmeans.predict(scaled)
```

```python
clusters
```

```python
# hitung jumlah data permasing2 cluster
unique_elements, counts_elements = np.unique(clusters, return_counts=True)
print("Frequency of unique values of the said array:")
print(np.asarray((unique_elements, counts_elements)))
```

```python
# baca data csv result_predict yang didapatkan dari cluster_training
df_result_predict = pd.read_csv('result_predict.csv')
```

```python
df_result_predict.head()
```

```python
# lengketkan hasil cluster diatas kedalam df_result_predict dengan nama kolom 'test'
df_result_predict['test'] = clusters
```
