```python
import pandas as pd
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.externals import joblib
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df_cluster = pd.read_csv('cluster_1.csv')
```

```python
df_cluster.head()
```

```python
print(df_cluster.shape)
```

```python
df_cluster = df_cluster.drop(['CARDNUM', 'cluster'], axis=1)
```

```python
print(df_cluster.shape)
```

```python
df_cluster.head()
```

```python
# convert pandas dataframe to list and then to one hot encoding
# cluster 1
records = []
for i in range(0, df_cluster.shape[0]):
    records.append([str(df_cluster.values[i,j]) for j in range(0, df_cluster.shape[1])])
    
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df_cluster_transform = pd.DataFrame(te_ary, columns=te.columns_)
df_cluster_transform.drop(['nan'], axis=1, inplace=True)
df_cluster_transform
```

```python
# Build up the frequent items# Build  
frequent_itemsets = apriori(df_cluster_transform, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemsets.shape
```

```python
frequent_itemsets.tail()
```

```python
frequent_itemsets.to_csv('frequent_itemsets_cluster_1.csv', index=False)
```

```python
# Create the rules# Create
# metric bisa diisi dengan 'support', 'confidence' atau 'lift'
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.02)
```
